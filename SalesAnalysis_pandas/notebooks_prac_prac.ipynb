```python
import pandas as pd
import os 
```

```python
#merging 12 months files

# df.head()

files = [file for file in os.listdir("./Sales_Data/")]
# print(files)
all_data = pd.DataFrame()
# print(all_data)
for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_data = pd.concat([all_data, df])

# print(all_data)
all_data.to_csv("all_data.csv", index = False)
```

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
# clean data
# drop nan rows
nan_df = all_data[all_data.isna().any(axis = 1)]
nan_df.head()
all_data = all_data.dropna(how = "all")
all_data.head()
```

```python
# ValueError: invalid literal for int() with base 10: 'Or'
# find and delete "or"
# temp_data = all_data[all_data["Order Date"].str[0:2]=="Or"]
# temp_data.head()
all_data = all_data[all_data["Order Date"].str[0:2] != "Or"]
# temp_data.head()
```

```python
# augment data with additional data
# add month column
all_data["Month"] = all_data["Order Date"].str[0:2]
all_data["Month"] = all_data["Month"].astype("int32")
all_data.head()
```

```python
# TypeError: can't multiply sequence by non-int of type 'str'
# convert columns to correct type
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"])
all_data["Price Each"] = pd.to_numeric(all_data["Price Each"])
all_data.head()
```

```python
#add sales data column
all_data["Sales"] = all_data["Quantity Ordered"]*all_data["Price Each"]
all_data.head()
```

```python
# add a city column
# all_data["City"] = all_data["Purchase Address"].apply(lambda x:x.split(",")[1]) #one option
def get_city(address):
    return address.split(",")[1]
all_data["City"] = all_data["Purchase Address"].apply(lambda x:get_city(x))
all_data.head()
```

```python
# question: what is the best month for sales?
all_data.groupby("Month").sum()
```

```python
# question: what city has heighest no. of sales?
all_data.groupby("City").sum()
```

```python
# what time should we display advertisements to maximise the liklihood of customer's bying product?

all_data["Order Date"] = pd.to_datetime(all_data["Order Date"])
all_data.head()
```

```python
all_data["Hour"] = all_data["Order Date"].dt.hour
all_data["Minutes"] = all_data["Order Date"].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby("Hour")]
all_data.groupby("Hour").count()
```

```python
# question: what products are often sold together

df = all_data[all_data["Order ID"].duplicated(keep = False)]
df["Grouped"] = df.groupby("Order ID")["Product"].transform(lambda x: ",".join(x))
df.head()
```

```python
df = df[["Order ID", "Grouped"]].drop_duplicates()
df.head()
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```
